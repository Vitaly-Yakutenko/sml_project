# Train

```python
items = []
with open('data/train.txt', 'rt') as f:
    line = f.readline()
    while line:
        numbers = line.split('\t')
        source = int(numbers[0])
        for sink in numbers[1:]:
            items.append((source, int(sink))) 
        line = f.readline()
len(items)
```

```python
data = pd.DataFrame(items, columns=['source', 'sink'])
```
